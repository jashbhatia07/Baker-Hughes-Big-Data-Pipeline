```python
import pandas as pd
import os
```

```python
site_df = pd.read_csv('./bde_data/site_metadata.csv')
site_df.head()
```

```python
engine_df = pd.read_csv('./bde_data/engine_metadata.csv')
engine_df.head()
```

```python
data_list = os.listdir('./bde_data/plant_data')
for file in data_list:
    data_frame = pd.read_csv(f'./bde_data/plant_data/{file}')
    data_frame.fillna(0, inplace=True)
    data_frame.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
    data_frame['TIME'] = pd.to_datetime(data_frame['DATE_TIME']).dt.time
    data_frame['DATE'] = pd.to_datetime(data_frame['DATE_TIME']).dt.date
    data_frame.drop(columns=['DATE_TIME'], inplace=True)
    for i in range(1, int((data_frame.shape[1]-3)/4)+1):
        for j in range(45):
            if site_df["PLANT_NAME"][j] == str(file)[:-4]:
                LHV = site_df["FUEL_LHV"][j]
                data_frame["THERMAL_EFF"] = (
                    data_frame["POWER_" + str(i)]/(data_frame["FUEL_FLOW_" + str(i)]*LHV))*100
    data_frame.to_csv(f'./analysed_data/' + file, index=False)
```
